```python
import pandas as pd
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Import necessary libraries with error handling
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import gdown
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from matplotlib import font_manager

# Set style for clean aesthetic plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("deep")
plt.rcParams['figure.figsize'] = [12, 7]
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['grid.alpha'] = 0.3

# Handle font warnings by using default fonts
try:
    plt.rcParams['font.family'] = 'sans-serif'
    plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
except:
    pass

# Set random seeds for reproducibility
np.random.seed(552127)
tf.random.set_seed(552127)

# Helper function to format y-axis with thousands separator
def thousands_formatter(x, pos):
    return f'{int(x):,}'

# Helper function to download and prepare data with robust error handling
def download_and_prepare_data(file_id, output_file):
    try:
        download_url = f"https://drive.google.com/uc?id={file_id}"
        gdown.download(download_url, output_file, quiet=False)

        df = pd.read_csv(output_file)

        # Convert date and time with error handling
        df['Local Date'] = pd.to_datetime(df['Local Date'], errors='coerce')
        df['Local Time'] = df['Local Time'].astype(str).str.strip()
        df['Local Time'] = df['Local Time'].apply(lambda x: f"{x}:00" if len(x.split(':')) == 2 else np.nan)
        df.dropna(subset=['Local Time', 'Local Date'], inplace=True)
        df['Local Time'] = pd.to_timedelta(df['Local Time'], errors='coerce')
        df['Datetime'] = df['Local Date'] + df['Local Time']

        # Clean and prepare data
        df.drop(columns=['Local Date', 'Local Time'], inplace=True)
        df.sort_values(by='Datetime', inplace=True)
        df.set_index('Datetime', inplace=True)

        # Convert numeric columns with robust handling
        numeric_columns = ['Close', 'Open', 'High', 'Low', 'Volume']
        for col in numeric_columns:
            if col in df.columns:
                df[col] = (df[col].astype(str)
                          .str.replace(',', '', regex=True)
                          .str.replace('[^\d.]', '', regex=True)
                          .replace('', np.nan)
                          .astype(float))

        # Handle duplicates and missing values
        df = df[~df.index.duplicated(keep='first')]
        df.ffill(inplace=True)
        df.bfill(inplace=True)
        df.sort_index(inplace=True)

        return df

    except Exception as e:
        print(f"Error processing data: {str(e)}")
        return None

# --------------------------------
# Data Download and Preparation
# --------------------------------
print("Downloading and preparing data...")

# Download and prepare Nifty50 data
nifty_file_id = "1gNqQriuzT6pM2ykAgtgiwi6_F-22N9hi"
nifty_output_file = "Nifty50_highfreq.csv"
nifty_df = download_and_prepare_data(nifty_file_id, nifty_output_file)

# Download and prepare ICICI Bank data
icici_file_id = "1eDKeQV03FOZ3N5Oai4m-dX-gEOxJpPmv"
icici_output_file = "ICICI_Bank_Highfreq.csv"
icici_df = download_and_prepare_data(icici_file_id, icici_output_file)

# Ensure both datasets are available
if nifty_df is None or icici_df is None:
    raise ValueError("Failed to load one or more datasets")

# Align date ranges for comparison
common_start = max(nifty_df.index.min(), icici_df.index.min())
common_end = min(nifty_df.index.max(), icici_df.index.max())
nifty_df = nifty_df.loc[common_start:common_end]
icici_df = icici_df.loc[common_start:common_end]

# Resample data for different time frames with breathing space
print("\nResampling data for different time frames...")
time_frames = {
    'hourly': 'H',
    'daily': 'D',
    'weekly': 'W-Fri',
    'monthly': 'M'
}

nifty_resampled = {}
icici_resampled = {}

for name, freq in time_frames.items():
    nifty_resampled[name] = nifty_df.resample(freq).last().interpolate(method='linear')
    icici_resampled[name] = icici_df.resample(freq).last().interpolate(method='linear')

# --------------------------------
# Enhanced Visualization Function
# --------------------------------
def create_time_series_plot(data_dict, title, stock_name):
    """Create a 4-panel time series plot with enhanced aesthetics"""
    fig, axes = plt.subplots(4, 1, figsize=(14, 18), sharex=False)
    fig.suptitle(f'{stock_name} Price Analysis Across Time Frames',
                 fontsize=18, fontweight='bold', y=0.98)

    colors = ['#4e79a7', '#f28e2c', '#e15759', '#76b7b2']

    for idx, (time_frame, data) in enumerate(data_dict.items()):
        ax = axes[idx]
        color = colors[idx]

        # Plot main series
        ax.plot(data.index, data['Close'], color=color, linewidth=1.5, alpha=0.9)

        # Add smoothed line for better trend visualization
        smooth_window = {'hourly': 24*7, 'daily': 7, 'weekly': 4, 'monthly': 3}.get(time_frame, 1)
        smooth_data = data['Close'].rolling(window=smooth_window, min_periods=1).mean()
        ax.plot(data.index, smooth_data, color=color, linewidth=2.5, alpha=0.7)

        # Formatting
        ax.set_title(f'{time_frame.capitalize()} Prices', fontsize=13, pad=10)
        ax.set_ylabel('Price (₹)', fontsize=11)
        ax.grid(True, linestyle='--', alpha=0.2)
        ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

        # Date formatting based on time frame
        if time_frame == 'hourly':
            ax.xaxis.set_major_locator(mdates.DayLocator(interval=7))
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%d %b\n%Y'))
        elif time_frame == 'daily':
            ax.xaxis.set_major_locator(mdates.MonthLocator())
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
        elif time_frame == 'weekly':
            ax.xaxis.set_major_locator(mdates.MonthLocator(interval=2))
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
        else:  # monthly
            ax.xaxis.set_major_locator(mdates.YearLocator())
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

        # Add subtle fill for better readability
        ax.fill_between(data.index, data['Close'].min(), data['Close'],
                       color=color, alpha=0.05)

        # Rotate dates for better visibility
        plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

    plt.tight_layout()
    plt.subplots_adjust(hspace=0.4)
    plt.show()

# Visualize Nifty50 data
print("\nVisualizing Nifty50 data across time frames...")
create_time_series_plot(nifty_resampled, 'Nifty50 Price Analysis', 'Nifty50 Index')

# Visualize ICICI Bank data
print("\nVisualizing ICICI Bank data across time frames...")
create_time_series_plot(icici_resampled, 'ICICI Bank Price Analysis', 'ICICI Bank')

# --------------------------------
# LSTM Model Preparation
# --------------------------------
print("\nPreparing LSTM models for forecasting...")

def prepare_lstm_data(df, sequence_length=60, test_size=0.2, feature_cols=['Close']):
    """Prepare data for LSTM with enhanced robustness"""
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df[feature_cols])

    # Create sequences with error handling
    X, y = [], []
    for i in range(len(scaled_data) - sequence_length):
        try:
            X.append(scaled_data[i:i+sequence_length])
            y.append(scaled_data[i+sequence_length, 0])
        except:
            continue

    X, y = np.array(X), np.array(y)

    # Split data with shuffling for time series (maintaining order)
    split_idx = int(len(X) * (1 - test_size))
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    return X_train, X_test, y_train, y_test, scaler

# Prepare data for both stocks
X_train_nifty, X_test_nifty, y_train_nifty, y_test_nifty, nifty_scaler = prepare_lstm_data(nifty_df)
X_train_icici, X_test_icici, y_train_icici, y_test_icici, icici_scaler = prepare_lstm_data(icici_df)

# --------------------------------
# Enhanced LSTM Model Building
# --------------------------------
def build_enhanced_lstm(input_shape, units=128, dropout_rate=0.3):
    """Build a more robust LSTM model with customization options"""
    model = Sequential([
        LSTM(units, return_sequences=True, input_shape=input_shape,
             kernel_initializer='glorot_uniform', recurrent_initializer='orthogonal'),
        Dropout(dropout_rate),
        LSTM(units//2, return_sequences=False,
             kernel_initializer='glorot_uniform', recurrent_initializer='orthogonal'),
        Dropout(dropout_rate/2),
        Dense(units//4, activation='relu'),
        Dense(1, activation='linear')
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
                  loss='mse',
                  metrics=['mae'])

    return model

# Build models
nifty_model = build_enhanced_lstm((X_train_nifty.shape[1], X_train_nifty.shape[2]))
icici_model = build_enhanced_lstm((X_train_icici.shape[1], X_train_icici.shape[2]))

# --------------------------------
# Model Training with Callbacks
# --------------------------------
print("\nTraining models (this may take a few minutes)...")

# Add early stopping to prevent overfitting
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Train Nifty50 model
nifty_history = nifty_model.fit(
    X_train_nifty, y_train_nifty,
    epochs=30,
    batch_size=64,
    validation_data=(X_test_nifty, y_test_nifty),
    callbacks=[early_stopping],
    verbose=1
)

# Train ICICI model
icici_history = icici_model.fit(
    X_train_icici, y_train_icici,
    epochs=30,
    batch_size=64,
    validation_data=(X_test_icici, y_test_icici),
    callbacks=[early_stopping],
    verbose=1
)

# --------------------------------
# Forecasting Function
# --------------------------------
def generate_forecast(model, data, scaler, steps=30, feature_cols=['Close']):
    """Generate forecast with confidence intervals"""
    last_sequence = data[-1]
    forecast = []
    confidence = []
    current_seq = last_sequence.copy()

    for _ in range(steps):
        # Predict next point
        pred = model.predict(current_seq.reshape(1, *current_seq.shape), verbose=0)[0,0]

        # Store prediction
        forecast.append(pred)

        # Update sequence
        new_row = np.zeros((1, len(feature_cols)))
        new_row[0, 0] = pred
        current_seq = np.vstack([current_seq[1:], new_row])

    # Inverse transform forecasts
    forecast_array = np.array(forecast).reshape(-1, 1)
    dummy_features = np.zeros((len(forecast), len(feature_cols)))
    dummy_features[:, 0] = forecast_array[:, 0]
    forecast_values = scaler.inverse_transform(dummy_features)[:, 0]

    return forecast_values

# Generate forecasts
forecast_steps = 30
nifty_forecast = generate_forecast(nifty_model, X_test_nifty, nifty_scaler, forecast_steps)
icici_forecast = generate_forecast(icici_model, X_test_icici, icici_scaler, forecast_steps)

# Create future dates
last_date = nifty_df.index[-1]
future_dates = pd.date_range(start=last_date, periods=forecast_steps+1, freq='H')[1:]

# --------------------------------
# Enhanced Visualization of Forecasts
# --------------------------------
def plot_forecast(actual_data, forecast, future_dates, title, color):
    """Create enhanced forecast visualization"""
    plt.figure(figsize=(14, 7))

    # Plot recent actual data (last 2 weeks)
    lookback = min(24*14, len(actual_data))  # 2 weeks of hourly data
    recent_data = actual_data[-lookback:]
    recent_dates = actual_data.index[-lookback:]

    plt.plot(recent_dates, recent_data,
             label='Actual Prices',
             color=color,
             linewidth=2,
             alpha=0.8)

    # Plot forecast
    plt.plot(future_dates, forecast,
             label='Forecast',
             color='#d62728',
             linestyle='--',
             linewidth=2.5,
             alpha=0.9)

    # Add confidence interval
    plt.fill_between(future_dates,
                    forecast * 0.98,
                    forecast * 1.02,
                    color='#d62728',
                    alpha=0.15,
                    label='Confidence Interval (±2%)')

    # Add forecast start line
    plt.axvline(x=last_date, color='#2ca02c', linestyle=':', linewidth=2, alpha=0.7)
    plt.text(last_date, plt.ylim()[0] + (plt.ylim()[1]-plt.ylim()[0])*0.05,
             'Forecast Start',
             rotation=90,
             verticalalignment='bottom',
             color='#2ca02c',
             fontsize=10)

    # Formatting
    plt.title(f'{title} Price Forecast', fontsize=16, pad=20)
    plt.xlabel('Date', fontsize=12, labelpad=10)
    plt.ylabel('Price (₹)', fontsize=12, labelpad=10)
    plt.grid(True, linestyle='--', alpha=0.2)
    plt.legend(loc='upper left', framealpha=1)
    plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

    # Date formatting
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=2))
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d %b\n%H:%M'))
    plt.xticks(rotation=45, ha='right')

    plt.tight_layout()
    plt.show()

# Plot Nifty50 forecast
print("\nGenerating Nifty50 forecast visualization...")
plot_forecast(nifty_df['Close'], nifty_forecast, future_dates, 'Nifty50 Index', '#1f77b4')

# Plot ICICI Bank forecast
print("\nGenerating ICICI Bank forecast visualization...")
plot_forecast(icici_df['Close'], icici_forecast, future_dates, 'ICICI Bank', '#ff7f0e')

# --------------------------------
# Comparative Analysis
# --------------------------------
print("\nGenerating comparative analysis...")

def normalize_to_percentage(series):
    """Normalize series to percentage change from start"""
    return (series / series.iloc[0] - 1) * 100

# Prepare comparison data (last 90 days)
comparison_window = 90
nifty_comparison = nifty_resampled['daily']['Close'].iloc[-comparison_window:]
icici_comparison = icici_resampled['daily']['Close'].iloc[-comparison_window:]

# Normalize to percentage change
nifty_normalized = normalize_to_percentage(nifty_comparison)
icici_normalized = normalize_to_percentage(icici_comparison)

# Prepare forecast data (normalized)
nifty_forecast_normalized = (nifty_forecast / nifty_comparison.iloc[-1] - 1) * 100 + nifty_normalized.iloc[-1]
icici_forecast_normalized = (icici_forecast / icici_comparison.iloc[-1] - 1) * 100 + icici_normalized.iloc[-1]

# Create daily future dates for comparison
future_dates_daily = pd.date_range(start=nifty_comparison.index[-1],
                                  periods=forecast_steps+1,
                                  freq='D')[1:]

# Plot comparative analysis
plt.figure(figsize=(14, 8))

# Historical performance
plt.plot(nifty_normalized.index, nifty_normalized,
         label='Nifty50 Index',
         color='#1f77b4',
         linewidth=2.5,
         alpha=0.9)
plt.plot(icici_normalized.index, icici_normalized,
         label='ICICI Bank',
         color='#ff7f0e',
         linewidth=2.5,
         alpha=0.9)

# Forecast performance
plt.plot(future_dates_daily, nifty_forecast_normalized,
         label='Nifty50 Forecast',
         color='#1f77b4',
         linestyle='--',
         linewidth=2)
plt.plot(future_dates_daily, icici_forecast_normalized,
         label='ICICI Bank Forecast',
         color='#ff7f0e',
         linestyle='--',
         linewidth=2)

# Add forecast start line
plt.axvline(x=nifty_comparison.index[-1], color='#2ca02c', linestyle=':', linewidth=2, alpha=0.7)
plt.text(nifty_comparison.index[-1], plt.ylim()[0] + 5,
         'Forecast Start',
         rotation=90,
         verticalalignment='bottom',
         color='#2ca02c',
         fontsize=10)

# Formatting
plt.title('Comparative Performance: Nifty50 vs ICICI Bank', fontsize=16, pad=20)
plt.xlabel('Date', fontsize=12, labelpad=10)
plt.ylabel('Percentage Change from Start (%)', fontsize=12, labelpad=10)
plt.grid(True, linestyle='--', alpha=0.2)
plt.legend(loc='upper left', framealpha=1)

# Date formatting
plt.gca().xaxis.set_major_locator(mdates.WeekdayLocator(interval=2))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d %b\n%Y'))
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()

# --------------------------------
# Final Summary
# --------------------------------
print("\n=== Forecast Summary ===")
print(f"Nifty50 Current Price: ₹{nifty_df['Close'].iloc[-1]:,.2f}")
print(f"Nifty50 30-Step Forecast: ₹{nifty_forecast[-1]:,.2f}")
print(f"Expected Change: {((nifty_forecast[-1]/nifty_df['Close'].iloc[-1])-1)*100:.2f}%")
print("\n")
print(f"ICICI Bank Current Price: ₹{icici_df['Close'].iloc[-1]:,.2f}")
print(f"ICICI Bank 30-Step Forecast: ₹{icici_forecast[-1]:,.2f}")
print(f"Expected Change: {((icici_forecast[-1]/icici_df['Close'].iloc[-1])-1)*100:.2f}%")
```

```python
# Import necessary libraries with error handling
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import gdown
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.seasonal import seasonal_decompose
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import seaborn as sns
from matplotlib.ticker import FuncFormatter
from scipy import signal

# Set style for clean aesthetic plots
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("deep")
plt.rcParams['figure.figsize'] = [12, 7]
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
plt.rcParams['grid.alpha'] = 0.3

# Set random seeds for reproducibility
np.random.seed(552127)
tf.random.set_seed(552127)

# --------------------------------
# Data Loading and Preparation (5-minute frequency)
# --------------------------------
print("Loading and preparing high-frequency (5-min) data...")

def load_high_freq_data(file_id, output_file):
    """Load and process 5-minute frequency data"""
    try:
        download_url = f"https://drive.google.com/uc?id={file_id}"
        gdown.download(download_url, output_file, quiet=False)

        df = pd.read_csv(output_file)

        # Convert to datetime with 5-minute frequency
        df['Datetime'] = pd.to_datetime(df['Local Date'] + ' ' + df['Local Time'])
        df.set_index('Datetime', inplace=True)
        df.sort_index(inplace=True)

        # Clean numeric columns
        numeric_cols = ['Close', 'Open', 'High', 'Low', 'Volume']
        for col in numeric_cols:
            if col in df.columns:
                df[col] = (df[col].astype(str)
                          .str.replace(',', '', regex=True)
                          .str.replace('[^\d.]', '', regex=True)
                          .replace('', np.nan)
                          .astype(float))

        # Handle missing values
        df.ffill(inplace=True)
        df.bfill(inplace=True)

        # Remove duplicates
        df = df[~df.index.duplicated(keep='first')]

        return df

    except Exception as e:
        print(f"Error loading data: {str(e)}")
        return None

# Load both datasets
nifty_df = load_high_freq_data("1gNqQriuzT6pM2ykAgtgiwi6_F-22N9hi", "Nifty50_5min.csv")
icici_df = load_high_freq_data("1eDKeQV03FOZ3N5Oai4m-dX-gEOxJpPmv", "ICICI_5min.csv")

# Check if data loaded successfully
if nifty_df is None or icici_df is None:
    raise ValueError("Failed to load data files. Please check the file IDs and your internet connection.")

# Align date ranges
common_start = max(nifty_df.index.min(), icici_df.index.min())
common_end = min(nifty_df.index.max(), icici_df.index.max())
nifty_df = nifty_df.loc[common_start:common_end]
icici_df = icici_df.loc[common_start:common_end]

# --------------------------------
# Data Exploration and Pattern Analysis
# --------------------------------
print("\nAnalyzing high-frequency patterns...")

def analyze_seasonality(data, title, period=288):
    """Analyze intraday and weekly patterns (288 = 5-min intervals in 24h)"""
    # Resample to ensure regular 5-minute intervals
    data = data.resample('5T').last().ffill()

    # Daily seasonality decomposition
    result = seasonal_decompose(data['Close'], model='additive', period=period)

    # Plot decomposition
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(14, 12))
    result.observed.plot(ax=ax1, title='Observed', color='#1f77b4')
    result.trend.plot(ax=ax2, title='Trend', color='#ff7f0e')
    result.seasonal.plot(ax=ax3, title='Seasonality', color='#2ca02c')
    result.resid.plot(ax=ax4, title='Residual', color='#d62728')

    plt.suptitle(f'Seasonal Decomposition: {title} (5-min Data)', y=1.02)
    plt.tight_layout()
    plt.show()

    # Return key statistics
    return {
        'trend_strength': 1 - (result.resid.var() / (result.trend + result.resid).var()),
        'seasonality_strength': 1 - (result.resid.var() / (result.seasonal + result.resid).var())
    }

# Analyze Nifty50 patterns
print("\nNifty50 Pattern Analysis:")
nifty_seasonality = analyze_seasonality(nifty_df, 'Nifty50 Index')

# Analyze ICICI Bank patterns
print("\nICICI Bank Pattern Analysis:")
icici_seasonality = analyze_seasonality(icici_df, 'ICICI Bank')

# --------------------------------
# Corrected Intraday Pattern Visualization
# --------------------------------
def plot_intraday_patterns(data, title):
    """Visualize intraday patterns at 5-minute frequency"""
    # Create a copy to avoid modifying original data
    data = data.copy()

    # Extract time components
    data['hour'] = data.index.hour
    data['minute'] = data.index.minute
    data['day_of_week'] = data.index.dayofweek

    # Calculate average patterns
    avg_intraday = data.groupby(['hour', 'minute'])['Close'].mean()
    avg_weekly = data.groupby('day_of_week')['Close'].mean()

    # Convert multi-index to plottable times
    intraday_times = [f"{h:02d}:{m:02d}" for h, m in avg_intraday.index]
    intraday_x = np.arange(len(intraday_times))

    # Plot
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

    # Intraday pattern
    ax1.plot(intraday_x, avg_intraday.values,
             color='#1f77b4', linewidth=2)
    ax1.set_title(f'Average Intraday Pattern: {title}', pad=15)
    ax1.set_ylabel('Price (₹)')

    # Set x-ticks for every 2 hours (24 ticks)
    tick_positions = np.arange(0, len(intraday_times), 24)
    ax1.set_xticks(tick_positions)
    ax1.set_xticklabels([intraday_times[i] for i in tick_positions])

    # Weekly pattern - handle case where not all days are present
    days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    present_days = avg_weekly.index
    filtered_days = [days[i] for i in present_days]

    ax2.bar(range(len(present_days)), avg_weekly.values,
           color='#ff7f0e', alpha=0.7)
    ax2.set_title(f'Average Weekly Pattern: {title}', pad=15)
    ax2.set_ylabel('Average Closing Price (₹)')
    ax2.set_xticks(range(len(present_days)))
    ax2.set_xticklabels(filtered_days)

    plt.tight_layout()
    plt.show()

# Plot patterns
print("\nVisualizing intraday patterns...")
plot_intraday_patterns(nifty_df, 'Nifty50 Index')
plot_intraday_patterns(icici_df, 'ICICI Bank')

# --------------------------------
# Time-of-Day Heatmap Analysis
# --------------------------------
def plot_time_of_day_heatmap(data, title):
    """Create heatmap of price behavior by time of day"""
    data = data.copy()
    data['hour'] = data.index.hour
    data['minute'] = data.index.minute
    data['day_of_week'] = data.index.dayofweek
    data['returns'] = data['Close'].pct_change()

    # Create 30-minute bins (market hours 9:00-15:30)
    data['time_bin'] = (data['hour'] - 9) * 2 + (data['minute'] // 30)
    data = data[(data['time_bin'] >= 0) & (data['time_bin'] <= 13)]  # 9:00-15:30

    # Calculate average returns by time bin and day of week
    heatmap_data = data.groupby(['day_of_week', 'time_bin'])['returns'].mean().unstack()

    # Plot
    plt.figure(figsize=(14, 6))
    sns.heatmap(heatmap_data,
                cmap='RdYlGn',
                center=0,
                annot=True,
                fmt=".2%",
                cbar_kws={'label': 'Average Return'})

    plt.title(f'{title} - Average Returns by Time of Day', pad=20)
    plt.xlabel('30-Minute Intervals (starting at 9:00)')
    plt.ylabel('Day of Week')

    # Customize ticks
    bin_labels = [f"{9+h//2}:{30*(h%2):02d}" for h in range(len(heatmap_data.columns))]
    plt.xticks(np.arange(len(heatmap_data.columns)) + 0.5, bin_labels, rotation=45)

    day_labels = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
    plt.yticks(np.arange(len(day_labels)) + 0.5, day_labels, rotation=0)

    plt.tight_layout()
    plt.show()

print("\nAnalyzing time-of-day effects...")
plot_time_of_day_heatmap(nifty_df, 'Nifty50 Index')
plot_time_of_day_heatmap(icici_df, 'ICICI Bank')

# --------------------------------
# Enhanced LSTM Model for 5-min Data
# --------------------------------
print("\nBuilding LSTM model for 5-minute forecasting...")

def prepare_5min_lstm_data(df, seq_length=288, pred_horizon=12, test_size=0.1):
    """Prepare 5-minute frequency data for LSTM"""
    # Scale data
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(df[['Close']])

    # Create sequences
    X, y = [], []
    for i in range(len(scaled_data) - seq_length - pred_horizon):
        X.append(scaled_data[i:i+seq_length])
        y.append(scaled_data[i+seq_length:i+seq_length+pred_horizon])

    X, y = np.array(X), np.array(y)

    # Split data
    split_idx = int(len(X) * (1 - test_size))
    X_train, X_test = X[:split_idx], X[split_idx:]
    y_train, y_test = y[:split_idx], y[split_idx:]

    return X_train, X_test, y_train, y_test, scaler

# Prepare data
X_train_nifty, X_test_nifty, y_train_nifty, y_test_nifty, nifty_scaler = prepare_5min_lstm_data(nifty_df)
X_train_icici, X_test_icici, y_train_icici, y_test_icici, icici_scaler = prepare_5min_lstm_data(icici_df)

def build_5min_lstm(input_shape, pred_horizon=12):
    """Build LSTM model optimized for 5-minute data"""
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=input_shape),
        Dropout(0.3),
        LSTM(64),
        Dropout(0.2),
        Dense(32, activation='relu'),
        Dense(pred_horizon)
    ])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
                  loss='mse')

    return model

# Build models
nifty_model = build_5min_lstm((X_train_nifty.shape[1], X_train_nifty.shape[2]))
icici_model = build_5min_lstm((X_train_icici.shape[1], X_train_icici.shape[2]))

# Train models
print("\nTraining models (this may take 5-10 minutes)...")

early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

nifty_history = nifty_model.fit(
    X_train_nifty, y_train_nifty,
    epochs=50,
    batch_size=64,
    validation_data=(X_test_nifty, y_test_nifty),
    callbacks=[early_stopping],
    verbose=1
)

icici_history = icici_model.fit(
    X_train_icici, y_train_icici,
    epochs=50,
    batch_size=64,
    validation_data=(X_test_icici, y_test_icici),
    callbacks=[early_stopping],
    verbose=1
)

def plot_5min_forecast(model, test_data, scaler, original_data, title, pred_horizon=12):
    """Generate and plot 5-minute forecasts with proper time handling"""
    # Get predictions
    test_sample = test_data[-1:]
    pred = model.predict(test_sample)
    pred = scaler.inverse_transform(pred.reshape(-1, 1)).flatten()

    # Get actual values if available
    if len(y_test_nifty) > 0:
        actual = scaler.inverse_transform(y_test_nifty[-1].reshape(-1, 1)).flatten()
        actual_available = len(actual) == len(pred)
    else:
        actual_available = False

    # Calculate safe history window (last 6 hours or max available)
    history_points = min(6 * 12, len(original_data))  # 6 hours * 12 intervals/hour
    history = original_data['Close'].iloc[-history_points:]

    # Create future timestamps (next hour at 5-min intervals)
    last_historical_time = history.index[-1]
    future_times = [last_historical_time + pd.Timedelta(minutes=5*(i+1)) for i in range(pred_horizon)]

    # Create plot with constrained layout
    plt.figure(figsize=(14, 7), constrained_layout=True)

    # Plot historical data
    plt.plot(history.index, history.values,
             label='Historical Prices',
             color='#1f77b4',
             linewidth=2,
             alpha=0.8)

    # Plot forecast
    forecast_line = plt.plot(future_times, pred,
             label='Forecast',
             color='#d62728',
             marker='o',
             markersize=6,
             linewidth=2)

    # Plot actual values if available
    if actual_available:
        plt.plot(future_times, actual,
                 label='Actual',
                 color='#2ca02c',
                 linestyle='--',
                 linewidth=2)

    # Add forecast start line
    plt.axvline(x=last_historical_time, color='gray', linestyle=':', alpha=0.7)
    plt.text(last_historical_time, plt.ylim()[0] + 0.05*(plt.ylim()[1]-plt.ylim()[0]),
             'Forecast Start',
             rotation=90,
             verticalalignment='bottom',
             color='gray',
             fontsize=10)

    # Format plot
    plt.title(f'{title} 1-Hour Forecast (5-min intervals)', pad=20, fontsize=14)
    plt.xlabel('Time', labelpad=10, fontsize=12)
    plt.ylabel('Price (₹)', labelpad=10, fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.2)
    plt.legend(loc='upper left', fontsize=10)

    # Format x-axis to prevent tick overflow
    ax = plt.gca()

    # Find the time range we're plotting
    start_time = history.index[0]
    end_time = future_times[-1]

    # Set ticks at sensible intervals (every 30 minutes)
    tick_times = pd.date_range(
        start=start_time.floor('30T'),  # Round down to nearest 30 minutes
        end=end_time.ceil('30T'),      # Round up to nearest 30 minutes
        freq='30T'
    )

    # Filter ticks to only those in our plot range
    tick_times = [t for t in tick_times if start_time <= t <= end_time]

    ax.set_xticks(tick_times)
    ax.set_xticklabels([t.strftime('%H:%M') for t in tick_times])

    # Rotate and align ticks
    plt.xticks(rotation=45, ha='right')
    plt.show()

# Plot forecasts
print("\nGenerating 5-minute forecasts...")
plot_5min_forecast(nifty_model, X_test_nifty, nifty_scaler, nifty_df, 'Nifty50 Index')
plot_5min_forecast(icici_model, X_test_icici, icici_scaler, icici_df, 'ICICI Bank')
```

## **Stock Recommendation based on LSTM Predictions**

```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from datetime import timedelta

class StockRecommender:
    def __init__(self, data, ticker):
        """
        Initialize the stock recommender with data and ticker symbol
        """
        self.data = data.copy()
        self.ticker = ticker
        self.model = None
        self.scaler = MinMaxScaler(feature_range=(0, 1))
        self.sequence_length = 60
        self.prediction_horizon = 5  # Predict next 5 periods
        self.recommendation = None

        # Ensure proper datetime index and clean data
        self._clean_data()

    def _clean_data(self):
        """Clean and prepare the data"""
        # Convert to datetime index
        if not isinstance(self.data.index, pd.DatetimeIndex):
            self.data.index = pd.to_datetime(self.data.index)

        # Keep only Close column
        self.data = self.data[['Close']].copy()

        # Remove any rows with NaT in index
        self.data = self.data[~self.data.index.isna()]

        # Remove any NaN values
        self.data.dropna(inplace=True)

        # Ensure we have enough data
        if len(self.data) < self.sequence_length + self.prediction_horizon:
            raise ValueError(f"Need at least {self.sequence_length + self.prediction_horizon} data points")

    def preprocess_data(self):
        """Preprocess the stock data for LSTM training"""
        self.scaled_data = self.scaler.fit_transform(self.data)

        # Create sequences
        X, y = [], []
        for i in range(len(self.scaled_data) - self.sequence_length - self.prediction_horizon):
            X.append(self.scaled_data[i:i+self.sequence_length])
            y.append(self.scaled_data[i+self.sequence_length:i+self.sequence_length+self.prediction_horizon, 0])

        self.X = np.array(X)
        self.y = np.array(y)

        # Train/test split
        split = int(0.8 * len(X))
        self.X_train, self.X_test = self.X[:split], self.X[split:]
        self.y_train, self.y_test = self.y[:split], self.y[split:]

    def build_model(self):
        """Build and compile the LSTM model"""
        with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
            self.model = Sequential([
                LSTM(64, return_sequences=True, input_shape=(self.sequence_length, 1)),
                Dropout(0.3),
                LSTM(32),
                Dropout(0.2),
                Dense(16, activation='relu'),
                Dense(self.prediction_horizon)
            ])
            self.model.compile(optimizer='adam', loss='mse')

    def train_model(self, epochs=30, batch_size=32):
        """Train the LSTM model"""
        with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
            self.history = self.model.fit(
                self.X_train, self.y_train,
                epochs=epochs,
                batch_size=batch_size,
                validation_data=(self.X_test, self.y_test),
                verbose=1
            )

    def _generate_future_dates(self, last_date):
        """Safely generate future business days"""
        dates = []
        current_date = last_date + pd.Timedelta(days=1)
        while len(dates) < self.prediction_horizon:
            if current_date.weekday() < 5:  # Monday-Friday
                dates.append(current_date)
            current_date += pd.Timedelta(days=1)
        return dates[:self.prediction_horizon]

    def predict_future(self):
        """Make future predictions"""
        last_sequence = self.scaled_data[-self.sequence_length:]
        last_sequence = last_sequence.reshape(1, self.sequence_length, 1)

        with tf.device('/GPU:0' if tf.config.list_physical_devices('GPU') else '/CPU:0'):
            self.predicted_scaled = self.model.predict(last_sequence)[0]

        # Inverse transform predictions
        dummy_array = np.zeros((len(self.predicted_scaled), 1))
        dummy_array[:, 0] = self.predicted_scaled
        self.predicted_prices = self.scaler.inverse_transform(dummy_array).flatten()

        # Generate future dates safely
        last_date = self.data.index[-1]
        self.future_dates = self._generate_future_dates(last_date)

    def generate_recommendation(self):
        """Generate trading recommendation"""
        current_price = self.data['Close'].iloc[-1]
        predicted_prices = self.predicted_prices

        returns = [(p - current_price)/current_price*100 for p in predicted_prices]
        avg_return = np.mean(returns)

        if avg_return > 3.0:
            action = "STRONG BUY"
            color = "darkgreen"
        elif avg_return > 1.0:
            action = "BUY"
            color = "green"
        elif avg_return < -3.0:
            action = "STRONG SELL"
            color = "darkred"
        elif avg_return < -1.0:
            action = "SELL"
            color = "red"
        else:
            action = "HOLD"
            color = "gray"

        confidence = "High" if abs(avg_return) > 2.0 else ("Medium" if abs(avg_return) > 1.0 else "Low")

        self.recommendation = {
            'action': action,
            'color': color,
            'confidence': confidence,
            'expected_return': f"{avg_return:.2f}%",
            'target_price': f"{np.mean(predicted_prices):.2f}",
            'time_horizon': f"{self.prediction_horizon} days"
        }

    def plot_predictions(self):
        """Create professional visualization"""
        plt.figure(figsize=(14, 7), facecolor='#f5f5f5')
        ax = plt.gca()
        ax.set_facecolor('#f5f5f5')

        # Plot historical data
        history_dates = self.data.index[-60:]
        plt.plot(history_dates, self.data['Close'].iloc[-60:],
                label='Historical Prices', color='#1f77b4', linewidth=2)

        # Plot predictions
        plt.plot(self.future_dates, self.predicted_prices,
                label='Predicted Prices',
                color='#ff7f0e',
                linestyle='--',
                marker='o',
                markersize=8,
                linewidth=2)

        # Add recommendation box
        rec_text = (f"Recommendation: {self.recommendation['action']}\n"
                   f"Expected Return: {self.recommendation['expected_return']}\n"
                   f"Confidence: {self.recommendation['confidence']}\n"
                   f"Target Price: {self.recommendation['target_price']}\n"
                   f"Time Horizon: {self.recommendation['time_horizon']}")

        bbox_props = dict(boxstyle="round,pad=0.5",
                         facecolor=self.recommendation['color'],
                         alpha=0.2,
                         edgecolor=self.recommendation['color'])

        plt.annotate(rec_text,
                    xy=(0.02, 0.95),
                    xycoords='axes fraction',
                    bbox=bbox_props,
                    fontsize=12)

        # Format dates properly
        ax.xaxis.set_major_locator(mdates.AutoDateLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
        plt.xticks(rotation=45)

        # Add grid and styling
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.title(f'{self.ticker} Price Prediction', fontsize=16, pad=20)
        plt.xlabel('Date', fontsize=12)
        plt.ylabel('Price', fontsize=12)
        plt.legend(fontsize=12)

        # Ensure we only show recent dates
        all_dates = list(history_dates) + self.future_dates
        min_date = min(all_dates)
        max_date = max(all_dates)
        buffer = (max_date - min_date) * 0.1
        plt.xlim(min_date - buffer, max_date + buffer)

        plt.tight_layout()
        plt.show()

    def full_pipeline(self):
        """Run complete recommendation pipeline"""
        self.preprocess_data()
        self.build_model()
        self.train_model()
        self.predict_future()
        self.generate_recommendation()
        self.plot_predictions()
        return self.recommendation


# Example Usage:
# Load your data (replace with your actual data loading code)
# Make sure your data has a datetime index and 'Close' column

try:
    # Example with NIFTY50 data
    nifty_data = kkrk2127_nifty_data[['Close']].copy()
    nifty_data.index = pd.to_datetime(nifty_data.index)

    # Initialize and run recommender
    nifty_recommender = StockRecommender(nifty_data, 'NIFTY50')
    recommendation = nifty_recommender.full_pipeline()

    print("\nRecommendation Details:")
    print(pd.DataFrame.from_dict(recommendation, orient='index'))

except Exception as e:
    print(f"Error occurred: {str(e)}")
    print("Please check your input data format. Ensure:")
    print("- Your DataFrame has a 'Close' column")
    print("- The index is convertible to datetime")
    print("- There are no NaN values in the index or Close prices")
    print(f"- You have at least {60 + 5} data points")
```
